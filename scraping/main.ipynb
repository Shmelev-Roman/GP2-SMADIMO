```python
import requests
from bs4 import BeautifulSoup

results = []

def parse_m2_pages(start_page=1, end_page=5):
    base_url = "https://m2.ru/moskva/nedvizhimost/kupit-kvartiru/"
    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/88.0.4324.96 Safari/537.36"
        )
    }

    global results
    all_urls = []

    for page_num in range(start_page, end_page + 1):
        cur_urls = []
        url = f"{base_url}?pageNumber={page_num}"
        resp = requests.get(url, headers=headers)
        print(f"Страница {page_num}, статус: {resp.status_code}")

        if resp.status_code != 200:
            # Если получаем 4xx/5xx, можно пропустить или прервать
            print("Ошибка, пропускаем страницу.\n")
            continue

        soup = BeautifulSoup(resp.text, "html.parser")
        card_links = soup.select("a.LinkSnippet.LinkSnippet_hover")

        if not card_links:
            print("На этой странице нет объявлений или вернулась пустая выдача. Прекращаем цикл.")
            break

        for link in card_links:
            href = link.get("href")
            if href:
                if not href.startswith("http"):
                    href = "https://m2.ru" + href
                all_urls.append(href)
                cur_urls.append(href)
        results += cur_urls

    return all_urls


result_links = parse_m2_pages(1, 500)
print(f"\nВсего ссылок собрано: {len(result_links)}")
for ln in result_links:
    print(ln)
```
